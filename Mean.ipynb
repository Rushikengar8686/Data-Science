```python
import numpy as np
import pandas as pd
```

```python
arr = np.array([2,3,1,4,5,6])
arr
```

```python
# To calculate mean using manualy
# formula of mean is mean = sum of all Data/total no of data
sumOfAlldata = arr.sum()
totalNoOfData = len(arr)
result = sumOfAlldata/totalNoOfData
result
```

```python
# To calculate mean using numpy
np.mean(arr)
```

```python
# to calulcate mean using pandas
arr.mean()
```

```python
dataSet = pd.read_csv("titanic_dataset.csv")
dataSet.head(5)
```

```python
dataSet.tail(5)
```

```python
# calculate mean using numpy bassed on colunm Age
np.mean(dataSet["Age"])
```

```python
# calculate mean using pandas bassed on colunm Age
dataSet["Age"].mean()
```

```python
# usng this dataset to show the histo graph
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(x = "Age", data= dataSet)
plt.show()
```

```python
sns.histplot(x = "Age", data = dataSet,bins=[i for i in range(21,38,2)])
plt.show()
```

```python
# to Display mean in red line
mn = dataSet["Age"].mean()
plt.plot([mn for i in range(0,350,2)],[i for i in range(0,350,2)],c = "red")
plt.show()
```
